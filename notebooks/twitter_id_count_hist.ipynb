```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('fivethirtyeight')
sns.set_style("white", {'ytick.major.size': 10.0})
sns.set_context("poster", font_scale=1.1)
```

```python
countList = []
with open('../data/twitter_id_counts/part-00000', 'r') as f:
    for cnt,line in enumerate(f):
        countList.append(int(line.strip().split('\t')[-1]))
```

```python
print 'There are',len(countList),'unique users in total'
print 'The maximum number of check-ins within all users is',max(countList)
```

```python
plt.figure()
plt.title('Histogram Plot for Number of Check-ins with no Range Specified')
plt.hist(countList)
plt.show()
```

```python
plt.figure()
plt.title('Histogram Plot for Number of Check-ins within Range 0~100')
plt.hist(countList, range=[0,100])
plt.show()
```

```python
plt.figure()
plt.title('Histogram Plot for Number of Check-ins within Range 0~50')
plt.hist(countList, range=[0,50])
plt.show()
```
